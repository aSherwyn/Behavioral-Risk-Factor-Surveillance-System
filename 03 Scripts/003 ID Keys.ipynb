# 2. ID Dicionary

```python
#Making id_dict a dictionary

id_dict = dict(id_df.values)
```

```python
#Checking id_dict

id_dict
```

# 3. class_id

```python
#Building dataframe of all class_id used in the dataset

c_ids = pd.DataFrame(df['class_id'].unique(), columns = ['id'])
```

```python
#Mapping classes to class_id

c_ids['class'] = c_ids['id'] #duplicating ids to a new column
c_ids.replace({'class' : id_dict}, inplace=True) #mapping values from data dictionary
```

```python
#Sorting and checking results

c_ids = c_ids.sort_values('id')
c_ids
```
