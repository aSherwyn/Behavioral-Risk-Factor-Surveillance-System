# 3. Defining Dictionary

### Building dictionary dataframe

```python
#Creating a dataframe of ID variables and their meanings

#Starting the dataframe
df_IDS = pd.DataFrame(df.groupby('ClassId')['Class'].min())

#Adding additional IDs and meanings to the dataframe
df_IDS = pd.concat([df_IDS, pd.DataFrame(df.groupby('TopicId')['Topic'].min())])
df_IDS = pd.concat([df_IDS, pd.DataFrame(df.groupby('LocationID')['Locationdesc'].min())])
df_IDS = pd.concat([df_IDS, pd.DataFrame(df.groupby('BreakoutID')['Break_Out'].min())])
df_IDS = pd.concat([df_IDS, pd.DataFrame(df.groupby('BreakOutCategoryID')['Break_Out_Category'].min())])
df_IDS = pd.concat([df_IDS, pd.DataFrame(df.groupby('QuestionID')['Question'].min())])
df_IDS = pd.concat([df_IDS, pd.DataFrame(df.groupby('ResponseID')['Response'].min())])
```

```python
#Making dataframes of ID variables and their meanings

df_IDS = pd.DataFrame(df.groupby('ClassId')['Class'].min())
df_IDS.columns = ['definition']

temp1 = pd.DataFrame(df.groupby('TopicId')['Topic'].min())
temp1.columns = ['definition']

temp2 = pd.DataFrame(df.groupby('LocationID')['Locationdesc'].min())
temp2.columns = ['definition']

temp3 = pd.DataFrame(df.groupby('BreakoutID')['Break_Out'].min())
temp3.columns = ['definition']

temp4 = pd.DataFrame(df.groupby('BreakOutCategoryID')['Break_Out_Category'].min())
temp4.columns = ['definition']

temp5 = pd.DataFrame(df.groupby('QuestionID')['Question'].min())
temp5.columns = ['definition']

temp6 = pd.DataFrame(df.groupby('ResponseID')['Response'].min())
temp6.columns = ['definition']
```

```python
#Count of all entries

df_IDS.count() + temp1.count() + temp2.count() + temp3.count() + temp4.count() + temp5.count() + temp6.count()
```

```python
#Combining ID dataframes

df_IDS = pd.concat([df_IDS, temp1])
df_IDS = pd.concat([df_IDS, temp2])
df_IDS = pd.concat([df_IDS, temp3])
df_IDS = pd.concat([df_IDS, temp4])
df_IDS = pd.concat([df_IDS, temp5])
df_IDS = pd.concat([df_IDS, temp6])
```
